```python
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
array = [[97, 160],
         [361, 423]]
df_cm = pd.DataFrame(array, index = ["edible", "inedible"],
                  columns = ["edible", "inedible"])
plt.figure(figsize = (6,4))
plt.title("Transfer Model")
sn.heatmap(df_cm, annot=True,cmap='Oranges', fmt='g', vmin=100, vmax=700)
plt.xlabel("Predicted class");
plt.ylabel("True class");
```
